```python
#grouping by hour of the day and computing revenue.
df_profit_hours = df.groupby("Hour").apply(lambda x: x["Item_Price"].sum())

#formatting output dataframe
df_profit_hours = df_profit_hours.to_frame()
df_profit_hours = df_profit_hours.reset_index()
df_profit_hours = df_profit_hours.rename(columns={0:"Revenue"})

#sorting by highest to lowest revenue generated
df_profit_hours = df_profit_hours.sort_values(by=["Revenue"] ,ascending =False)
df_profit_hours = df_profit_hours.reset_index(drop=True)

#rounding revenue to integer
df_profit_hours["Revenue"] = df_profit_hours["Revenue"].astype(int)
df_profit_hours.head()
```

```python
#creating a unique code for each day representing the date of each weekday to count unique days of the week.
df["unique_day_code"] = df["Year"].astype(str) + df["Month"].astype(str) + df["Day"].astype(str)

#counting total number of weekdays in the data
df_day_count = df.groupby("Weekday")["unique_day_code"].nunique()

#code to format output
df_day_count= df_day_count.to_frame()
df_day_count = df_day_count.reset_index()
df_day_count = df_day_count.rename(columns={"unique_day_code":"Number of Days"})

#getting count of unique transactions for each weekday
df_trans_count = df.groupby("Weekday")["Transaction"].nunique()
df_trans_count = df_trans_count.to_frame()
df_trans_count = df_trans_count.reset_index()

#merging dataframes to have both transactions and number of days
df_q4 = pd.merge(df_day_count,df_trans_count,on="Weekday")

#Finding Number of Baristas required for each weekday.
df_q4["Average Transactions"] = df_q4["Transaction"]/df_q4["Number of Days"]
df_q4["Barista Count"] =  np.ceil(df_q4["Average Transactions"] / 50) 

df_q4[["Weekday","Barista Count"]].head(7)
```

```python
#loading item grouping csv that contains the mapping of each item to a group(Food,Drink or Unknown)
df2 = pd.read_csv("/content/drive/MyDrive/Data Science Eugene Pinsky/Homework_8/bakery dataset/BreadBasket_DMS_item_grouping.csv")
df2 = df2.rename(columns={"Item ":"Item"})

#Creating a merged dataframe with group column.
df3 = pd.merge(df,df2,on="Item")
df3 = df3.sort_values(by="Transaction")

#grouping by Item group and calcultaing mean/average price.
df_group_average = df3.groupby("Group")["Item_Price"].mean()

#formatting output dataframe
df_group_average = df_group_average.to_frame()
df_group_average = df_group_average.reset_index()

#rounding average price
df_group_average["Item_Price"] = df_group_average["Item_Price"].round(decimals=2)
df_group_average = df_group_average.rename(columns={"Item_Price":"Average Price"})
df_group_average.head(2)
```

```python
df3 = df3.reset_index(drop=True)

#Counting number of items in each group
df_q9_1 = df3.groupby("Group")["Group"].count()
df_q9_1 = df_q9_1.to_frame()
df_q9_1 = df_q9_1.rename(columns={"Group":"Total"})
df_q9_1 = df_q9_1.reset_index()

#Counting unique transactions for each group
df_q9_2 = df3.groupby("Group")["Transaction"].nunique()
df_q9_2 = df_q9_2.to_frame()
df_q9_2 = df_q9_2.rename(columns={"Transaction":"Transactions"})
df_q9_2 = df_q9_2.reset_index()

df_q9_3 = pd.merge(df_q9_1,df_q9_2,on="Group")

#Computing items per transaction for all groups
df_q9_3["Items per transaction"] = df_q9_3["Total"]/df_q9_3["Transactions"]
df_q9_3["Items per transaction"] = df_q9_3["Items per transaction"].apply(lambda x: math.ceil(x))
df_q9_3[["Group","Items per transaction"]].head(1)
```
