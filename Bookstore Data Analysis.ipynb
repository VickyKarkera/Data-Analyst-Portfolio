Q1. load the dataset csv file using Pandas.

```python
#importing pandas and reading using read csv function
import pandas as pd
df = pd.read_csv("/content/books_scraped.csv")
df.head()
```

Q2. compute the median price PM of a book.

```python
#using quantile and round functions to compute median and round the final result
Median_Price_0f_Books = df["Price"].quantile(0.5)
Median_Price_0f_Books = Median_Price_0f_Books.round(2)
print("The median price for the books are: "+str(Median_Price_0f_Books))
```

Q3. suppose we divide all books into 2 groups: ”cheap” books
with prices less than PM and ”expensive” books with prices
more than PM. What is the average price in each group and
what is the median price in group? Which one is larger for
each group.

```python
#Filtering records in pandas by price using median as criteria
Cheap_books = df[df["Price"] < Median_Price_0f_Books ]
Expensive_books = df[df["Price"] > Median_Price_0f_Books ]

#computing means for both groups and rounding results
Mean_Of_Cheap_Books = Cheap_books["Price"].mean()
Mean_Of_Cheap_Books = Mean_Of_Cheap_Books.round(2)

Mean_Of_Expensive_Books = Expensive_books["Price"].mean()
Mean_Of_Expensive_Books = Mean_Of_Expensive_Books.round(2)

#computing median for both groups and rounding results
Median_Of_Cheap_Books = Cheap_books["Price"].median()
Median_Of_Cheap_Books = Median_Of_Cheap_Books.round(2)

Median_Of_Expensive_Books = Expensive_books["Price"].median()
Median_Of_Expensive_Books = Median_Of_Expensive_Books.round(2)

print("----------------------------------")
print("| Type\t    | Median\t| Average|")
print("----------------------------------")
print("| Expensive |".rjust(11)+" "+str(Median_Of_Expensive_Books).rjust(2)+"\t| "+str(Mean_Of_Expensive_Books)+"  |")
print("|    Cheap  |".rjust(11)+" "+str(Median_Of_Cheap_Books).rjust(2)+"\t| "+str(Mean_Of_Cheap_Books)+"  |")
print("----------------------------------")

print("\nThe averages for both categories of books is higher.")
```

Q4. compute average price of a book by Star Rating.

```python
#using groupby pandas function to group by star rating and computing mean price for each group
df.groupby("Star_rating",as_index=False)["Price"].mean().round(2)
```

Q5. compute average price of a book by Book Category.

```python
#using groupby pandas function to group by book category and computing mean price for each group
df.groupby("Book_category",as_index=False)["Price"].mean().round(2)
```

Q6. compute average price of a book by Book Category and
Star Rating.

```python
#using groupby pandas function to group by book category and Star rating and finally computing mean price for each group
df.groupby(["Star_rating","Book_category"],as_index=False)["Price"].mean().round(2)
```

Q7. assume that Quantity reflects the demand. For each Book
Category, compute the average demand (round the result
to integers). What are the top 3 Book Categories by this
criteria?

```python
#using groupby pandas function to group by book category and Star rating computing mean quantity 
#( Demand ) for each group and rounding result to integers.
Demand_By_Category = df.groupby("Book_category")['Quantity'].agg(['mean']).round(0).astype(int)

#sorting book categories by demand (Average Quantity) and displaying first 13 as these represent 
#top 3 categories. 
Demand_By_Category.sort_values(by='mean',ascending=False).head(13)
```

The top 3 book categories grouped according to demand are illustrated in the table above in decreasing order of demand as we go from top to bottom.
